```python
import sys, pickletools, gzip, zipfile, io
p = "/home/nitesh/IsaacSim5/IsaacLab/offline_rl_models/flava/flava_iql_final_20251103_135553.pkl"

# a) Is it gzipped?
with open(p,'rb') as f:
    magic = f.read(2)
    print("Magic bytes:", magic)
    if magic == b'\x1f\x8b':
        print("Looks like gzip.")
    elif magic == b'PK':
        print("Looks like a ZIP/NPZ.")
    else:
        print("Not gzip/zip by magic bytes.")

# b) Try to disassemble as a pickle (will error if not)
try:
    pickletools.dis(open(p,'rb').read(2000))
except Exception as e:
    print("pickletools.dis error (expected if not a plain pickle):", e)
```

```python
import zipfile, os

with zipfile.ZipFile(p, "r") as zf:
    names = zf.namelist()
    print("ZIP entries:", names)
```
